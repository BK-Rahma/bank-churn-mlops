```python
#dans jupyter lab
import requests
import json

# URL de ton API FastAPI
url = "http://localhost:8000/predict"

# Données à envoyer
data = {
    "CreditScore": 650,
    "Age": 35,
    "Tenure": 5,
    "Balance": 50000,
    "NumOfProducts": 2,
    "HasCrCard": 1,
    "IsActiveMember": 1,
    "EstimatedSalary": 75000,
    "Geography_Germany": 0,
    "Geography_Spain": 1
}

# Envoyer la requête POST
response = requests.post(url, json=data)

# Afficher la réponse
print(f"Status Code: {response.status_code}")
print(f"Response: {response.json()}")
```
